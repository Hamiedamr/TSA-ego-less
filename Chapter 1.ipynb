# Introduction
## What is the time series?
> A sequence of data points organized in time order

- The sequence captures data at equally
    spaced points in time
- Data collected irregularly is not
    considered a time series


## Application of time series analysis
Time series analysis is used in the following:
- Economic forecasting
- Stock market analysis
- Demand planning and forecasting
- Anomaly detection

and much more...

## Better understanding of Time series constrction
we can say that time series consists of three main components:
1. Trend: Change over long period
2. Seasonality: Periodic behavior 
3. Residuals: Irregular fluctuations over time

### Decompostion Models
1. Additive
2. Multiplicative
3. Pseudo additive

## Additive Models
Assume that the observed time series is the sum of its
components
- Observation = Trend + Seasonality + Residual
- Additive models are used when the magnitudes of the seasonal and residual
    values are independent of trend

# Imports

```python
%matplotlib inline

import sys
import statsmodels as ss
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
```

```python
time = np.arange(1, 51)
```

Now we need to create a trend. Let's pretend we have a sensor measuring demand for electricity. We'll ignore units to keep things simple.

```python
trend = time * 2.75
```

What folllows is a plot showing trend as a function of time.

```python
fig, ax = plt.subplots()
ax.plot(time, trend, 'b')
ax.set_title("Trend vs Time")
ax.set_xlabel("minutes")
ax.set_ylabel("electricity demand");
```

Great! Now let's generate a seasonal component.

```python
seasonal = 10 + np.sin(time) * 10
```

Naturally, we'll want to plot our seasonality.

```python
fig, ax2 = plt.subplots()
ax2.plot(time, seasonal, 'g')
ax2.set_title("Seasonality vs Time")
ax2.set_xlabel("minutes")
ax2.set_ylabel("electricity demand")
```

And now to construct the residual component.

```python
np.random.seed(10)  # reproducible results
residual = np.random.normal(loc=0.0, scale=1, size=len(time))
```

A quick plot of residuals...

```python
fig, ax3 = plt.subplots()
ax3.plot(time, residual, 'r')
ax3.set_title("Residuals vs Time")
ax3.set_xlabel("minutes")
ax3.set_ylabel("electricity demand");
```

### Additive Time Series
$O_{t} = T_{t} + S_{t} + R_{t}$ 

$O_{t}$ is the output   
$T_{t}$ is the trend  
$S_{t}$ is the seasonality  
$R_{t}$ is the residual  
$_t$ is a variable representing a particular point in time 

```python
additive = trend + seasonal + residual
```

```python
plt.plot(time, additive, 'k')
plt.title("Additive Time Series")
plt.xlabel("minutes")
plt.ylabel("electricity demand");
```

## Multiplicative Models
assume that the observed time series is the product of its components
- Observation = Trend * Seasonality * Residual
- It is possible to transform a multiplicative model to an additive by applying a
    log transformation.
- Multiplicative models are used when the magnitudes of the seasonal and
    residual values fluctuate with trend

### Multiplicative Time Series

$O_{t} = T_{t} * S_{t} * R_{t}$ 

```python
multiplicative = trend * seasonal * residual
```

```python
fig, ax = plt.subplots()
ax.plot(time, multiplicative, 'k-.')
ax.set_title("Multiplicative Time Series")
ax.set_xlabel("minutes")
ax.set_ylabel("electricity demand");
```

```python
# ignoring residual to make pattern more apparent
ignored_residual = np.ones_like(residual)
multiplicative = trend * seasonal * ignored_residual
fig, ax = plt.subplots()
ax.plot(time, multiplicative, 'k-.')
ax.set_title("Multiplicative Time Series")
ax.set_xlabel("minutes")
ax.set_ylabel("electricity demand");
```

## Difference between additive and multiplicative models

```python
fig, axs = plt.subplots(1,2)
fig.set_figwidth(15)
axs[0].plot(additive, 'k.')
ax2 = axs[0].twinx()
ax2.plot(time, trend)
axs[0].set_title('Additive Model')
axs[1].plot(multiplicative, 'k-.')
ax2 = axs[1].twinx()
ax2.plot(time, trend)
axs[1].set_title('Multiplicative Model')
```

# Time Series Decomposition

Now let's reverse directions. We have additive and multiplicative data. Let's decompose them into their three components. 
> *look [here](http://www.statsmodels.org/dev/generated/statsmodels.tsa.seasonal.seasonal_decompose.html) for details on how the following method works.*

### Additive Decomposition

```python
from statsmodels.tsa.seasonal import seasonal_decompose
x = pd.Series(additive)
ss_decomposition = seasonal_decompose(x=x, model='additive', period=10)
estimated_trend = ss_decomposition.trend
estimated_seasonal = ss_decomposition.seasonal
estimated_residual = ss_decomposition.resid
```

### Additive Plots

```python
fig, axes = plt.subplots(4, 1, sharex=True, sharey=False)
fig.set_figheight(10)
fig.set_figwidth(15)

axes[0].plot(additive, label='Original')
axes[0].legend(loc='upper left');

axes[1].plot(estimated_trend, label='Trend')
axes[1].legend(loc='upper left');

axes[2].plot(estimated_seasonal, label='Seasonality')
axes[2].legend(loc='upper left');

axes[3].plot(estimated_residual, label='Residuals')
axes[3].legend(loc='upper left');
```

### Multiplicative Decomposition

```python
ss_decomposition = seasonal_decompose(x=multiplicative, model='multiplicative', period=10)
estimated_trend = ss_decomposition.trend
estimated_seasonal = ss_decomposition.seasonal
estimated_residual = ss_decomposition.resid
```

### Multiplicative Plots

```python
fig, axes = plt.subplots(4, 1, sharex=True, sharey=False)
fig.set_figheight(10)
fig.set_figwidth(15)

axes[0].plot(multiplicative, label='Original')
axes[0].legend(loc='upper left');

axes[1].plot(estimated_trend, label='Trend')
axes[1].legend(loc='upper left');

axes[2].plot(estimated_seasonal, label='Seasonality')
axes[2].legend(loc='upper left');

axes[3].plot(estimated_residual, label='Residuals')
axes[3].legend(loc='upper left');
```
